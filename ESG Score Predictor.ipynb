```python
import yfinance as yf
```

```python
def getStock(TICKER):
  START_DATE = '2020-01-01' # Stock data start date
  END_DATE = '2023-09-01' # Stock data end date

  stockPx = yf.download(TICKER, START_DATE)
  return stockPx

def getIS(TICKER):
    stock = yf.Ticker(TICKER)
    return stock.income_stmt.T.head(1)

def getBS(TICKER):
    stock = yf.Ticker(TICKER)
    return stock.balance_sheet.T.head(1)

def getCS(TICKER):
    stock = yf.Ticker(TICKER)
    return stock.cash_flow.T.head(1)
```

```python
def getCompleteStatement(TICKER):
  df_bs = getBS(TICKER)
  df_cs = getCS(TICKER)
  df_is = getIS(TICKER)
  df_combined = pd.concat([df_bs, df_cs, df_is], axis=1)
  return df_combined
```

```python
df_list = []

for i in df.index:
    try:
      small_df = pd.DataFrame(df.loc[i]).transpose()
      statement = getCompleteStatement(df["Symbol"][i])
      print(statement)

      statement.index = small_df.index

      combined_df = small_df.join(statement)
      df_list.append(combined_df)
    except:
      pass

big_df = pd.DataFrame().append(df_list)
```

```python
big_df = pd.concat(df_list, ignore_index=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plotHistoGraph(item):
  random_companies = np.random.choice(big_df['Symbol'], 50, replace=False)

  filtered_df = big_df[big_df['Symbol'].isin(big_df)]

  plt.figure(figsize=(10, 6))
  plt.bar(filtered_df['Symbol'], filtered_df[item], color='skyblue')
  plt.xlabel('Company Symbol')
  plt.ylabel(item)
  plt.xticks(rotation=80)
  plt.tight_layout()

  plt.show()
```

```python
def plotScatterGraph(item1, item2):
    random_companies = np.random.choice(big_df['Symbol'], 50, replace=False)

    filtered_df = big_df[big_df['Symbol'].isin(random_companies)]

    plt.figure(figsize=(10, 6))
    plt.scatter(filtered_df[item1], filtered_df[item2], color='skyblue')
    plt.xlabel(item1)
    plt.ylabel(item2)
    plt.tight_layout()

    plt.show()
```

```python
def plotBox(item):
    random_companies = np.random.choice(big_df['Symbol'], 100, replace=False)
    filtered_df = big_df[big_df['Symbol'].isin(random_companies)]

    plt.figure(figsize=(6, 4))

    plt.boxplot(filtered_df[item])
    plt.xticks([1], [item])
    plt.xlabel('Items')
    plt.ylabel('Values')
    plt.title('Box Plot for ' + item)

    plt.tight_layout()
    plt.show()
```
